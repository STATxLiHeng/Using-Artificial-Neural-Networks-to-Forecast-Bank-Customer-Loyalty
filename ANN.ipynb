# Data Preprocessing

```python
dataset = pd.read_csv('../Deep Learning Data Udemy/ANN/Churn_Modelling.csv')
dataset.head()
```

# Create Feature Matrix & Dependence Variable

```python
# Feature Matrix
X = dataset.iloc[:,3:13].values
# Dependence
y = dataset.iloc[:,-1].values
X[0],y[0]
```

```python
X.shape
```

# Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
# For country
labelencoder_X_1 = LabelEncoder()
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])


# For sex
labelencoder_X_2 = LabelEncoder()
X[:,2] = labelencoder_X_2.fit_transform(X[:,2])


# OneHotEncoder
onehotencoder = ColumnTransformer([("encoder",OneHotEncoder(), [1])], remainder = 'passthrough')
X = onehotencoder.fit_transform(X)
# delete one dummy from countries
X = X[:,1:]
X[0]
```

# Splitting the dataset into the Training set and Testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = .2, random_state = 0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initializing  the ANN
classifier = Sequential()

# Add the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu',input_dim = 11))

# Add the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu'))

# Add the output layer
classifier.add(Dense(units = 1, kernel_initializer='uniform',activation='sigmoid'))
```

# Compile the ANN

```python
classifier.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

# Fitting the ANN to the Training Set

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

# Making the predictions and evaluating the model.

```python
# Predicting the Test set result
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
# Making Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

# Evaluating the ANN (Cross Validation)

```python
dataset = pd.read_csv('../Deep Learning Data Udemy/ANN/Churn_Modelling.csv')
#dataset.head()

# Feature Matrix
X = dataset.iloc[:,3:13].values
# Dependence
y = dataset.iloc[:,-1].values

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
# For country
labelencoder_X_1 = LabelEncoder()
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])


# For sex
labelencoder_X_2 = LabelEncoder()
X[:,2] = labelencoder_X_2.fit_transform(X[:,2])


# OneHotEncoder
onehotencoder = ColumnTransformer([("encoder",OneHotEncoder(), [1])], remainder = 'passthrough')
X = onehotencoder.fit_transform(X)
# delete one dummy from countries
X = X[:,1:]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = .2, random_state = 0)#

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Cross Validation
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
  # Initializing  the ANN
  classifier = Sequential()

  # Add the input layer and the first hidden layer
  classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu',input_dim = 11))

  # Add the second hidden layer
  classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu'))

  # Add the output layer
  classifier.add(Dense(units = 1, kernel_initializer='uniform',activation='sigmoid'))

  # Compile the ANN
  classifier.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])

  return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
# n_jobs為使用所有的CPU數量，-1為全部
accuracies =  cross_val_score(estimator = classifier, Ｘ = X_train, y = y_train, cv = 10, n_jobs = 3)
```

```python
mean = accuracies.mean()
variance = accuracies.std()
mean, variance
```

# Tunnung the ANN with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
def build_classifier(optimizer = 'adam'):
  # Initializing  the ANN
  classifier = Sequential()
  # Add the input layer and the first hidden layer
  classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu',input_dim = 11))
  # Add the second hidden layer
  classifier.add(Dense(units = 6, kernel_initializer='uniform',activation='relu'))
  # Add the output layer
  classifier.add(Dense(units = 1, kernel_initializer='uniform',activation='sigmoid'))
  # Compile the ANN
  classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])
  return classifier

classifier = KerasClassifier(build_fn = build_classifier)
```

```python
# Setting parameter combination
parameters = {'batch_size': [25, 32],
            'epochs': [100, 500],
            'optimizer': ['adam','rmsprop']}

grid_search = GridSearchCV(estimator = classifier,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 3)
```

```python
grid_search = grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
best_accurarcy = grid_search.best_score_
best_parameters, best_accurarcy
```
